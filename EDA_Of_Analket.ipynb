```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matches_sheet_id="14dFI_VPQRRZoEwOIt843k46fOP3LFwlodd_m9S_qoNY"
matches_sheet_name="Matches_After_Cleaning"

matches= pd.read_csv("https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(matches_sheet_id,matches_sheet_name))
```

```python
deliveries_sheet_id="1pYQ2GHb2qqkLwuwkDkuIu5rWmmuF_2sHtZvjmFLkNgs"
deliveries_sheet_name="Deliveries_After_Cleaning"

deliveries= pd.read_csv("https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(deliveries_sheet_id,deliveries_sheet_name))
```

```python
deliveries_matches_sheet_id="1Vw3qJN1pZzJj7DWLDiXQdCB8eGv3YQrMNmUsFJR3o9M"
deliveries_matches_sheet_name="Merged_Deliveries_and_Matches"

deliveries_matches= pd.read_csv("https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(deliveries_matches_sheet_id,deliveries_matches_sheet_name))
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(deliveries.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
corr = deliveries.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr, annot=True, square=True,linewidths=.5, ax=ax,cmap="YlGnBu")
plt.show()
```

```python
corr = deliveries_matches.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr, annot=True, square=True,linewidths=.5, ax=ax, cmap="YlGnBu")
plt.show()
```

```python
corr = matches.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr, annot=True, square=True,linewidths=.5, ax=ax,cmap="YlGnBu")
plt.show()
```

```python

plt.figure(figsize=(12,8))
ax=sns.countplot(y="toss_winner", data=matches, orient="h", hue="toss_decision",palette=['#432371',"#FAAE7B"])
plt.xlabel("Toss Decision",fontsize=14)
plt.ylabel("Teams",fontsize=14)
plt.title("Teams Toss Decision",fontsize=16)
```

```python
plt.figure(figsize=(12,8))
ax=sns.countplot(y="winner", data=matches, orient="h", hue="toss_decision",palette=['#432371',"#FAAE7B"])
plt.xlabel("Toss Decision",fontsize=14)
plt.ylabel("Teams",fontsize=14)
plt.title("Teams Toss Decision",fontsize=16)
```

```python
plt.figure(figsize=(12,8))
plt.grid(True)
sns.countplot(y=matches["winner"], orient="h",palette=["#800080","#FFFF00","#FF2400","#950217","#0063B2FF","#FF0000","#FF8C00","#83EEFF","#4B4A74","#CCACDB","#000000","#B5651E"])
plt.xlabel("Wins",fontsize=14)
plt.ylabel("Teams",fontsize=14)
plt.title("Teams Wins",fontsize=16)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(matches['winner'][matches['eliminator']=='Yes'],orient="h" ,data=matches)
plt.title('Most Wins in Super Over',fontsize=16)
plt.grid(True)
plt.xlabel('Teams',fontsize=14)
plt.ylabel('Wins',fontsize=14)
```

```python
plt.figure(figsize=(10,8))
player_of_match=matches['player_of_match'].value_counts()[:25].index
sns.countplot(order=player_of_match, data=matches, y="player_of_match",palette=["#800080","#FFFF00","#FF2400","#950217","#0063B2FF","#FF0000","#FF8C00","#83EEFF","#4B4A74","#CCACDB"])
plt.title("Top 25 Player of the Match",fontsize=16)
plt.xlabel('Frequents Player of the  Match Number', fontsize = 14)
plt.ylabel('Players', fontsize = 14)
```

```python
plt.figure(figsize=(20,20))
sns.displot(matches['result_margin'])
plt.grid(True)
plt.title("Result Margin",fontsize=16)
plt.xlabel('Runs/Wickets', fontsize = 14)
plt.ylabel('Frquency', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize=(15,15))
count_venue_number=matches['venue'].value_counts().head(10).values
count_venue_name=matches['venue'].value_counts().head(10).index
sns.barplot(x = count_venue_number, y=count_venue_name , data = matches)
plt.grid(True)
plt.title("Top 10 Stadiums Based on Number of Matches",fontsize=16)
plt.xlabel('Number of Matches', fontsize = 14)
plt.ylabel('Stadium Name', fontsize = 14)
```

```python
plt.figure(figsize=(15,15))
sns.countplot(data=deliveries[deliveries['dismissal_kind']!='none'], x='dismissal_kind')
plt.title('Dismissed Types',fontsize=16)
plt.xlabel("Dissmissals",fontsize=14)
plt.ylabel("Frequency",fontsize=14)
plt.grid(True)
plt.show()
```
